# Performing EDA on sample superstore data

```python
# importing all the required libraries
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# loading the data
data=pd.read_csv('SampleSuperstore.csv')
```

```python
# deleting the variable postal code as it is insignificant 
col=['Postal Code']
data=data.drop(columns=col,axis=1)
```

```python
data.info()
```

```python
# visualaizing the data by plots
```

```python
data.hist(bins=50 ,figsize=(20,15))
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(data['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=data['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(data['Region'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=data['Region'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(data['Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=data['Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
## office suppliers selled more
```

```python
sns.catplot("Ship Mode", hue="Segment", data=data, kind="count", aspect=1.5, palette="Set1")
```

```python
plt.figure(figsize=[12,10])
ax = sns.barplot(x="Region", y="Sales", hue="Segment", data=data, palette="Set1")
```

```python
data.corr()
```

```python
# correlation between the variables
sns.heatmap(data.corr(), annot= True)
plt.show()
```

```python
# sales and profits are highly correlated.
```

```python
# we will check how sales and profit vary each other
sns.pairplot(data)
```

```python
# statewise profit statistical information
data.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```
